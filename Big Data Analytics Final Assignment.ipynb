### Checking for outliers

```python
plt.boxplot(df['Profit'])
plt.show()
```

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers = df[(df['Profit'] < lower_bound) | (df['Profit'] > upper_bound)]
print(outliers['Profit'])
```

The outliers contains values of Profit that are significant for the analysis of sales. So the outliers will be kept.

### Now we can proceed for Visualization

```python
# There are inconsistencies in the month names.
# Mapping the incorrect month names to correct ones
month_map = {'Marsh': 'March', 'Aogust': 'August', 'Jone': 'June', 'Joly': 'July'}

# Replacing the incorrect month names with correct ones
df['Month'] = df['Month'].replace(month_map)

# Sorting the DataFrame by the month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
df = df.sort_values('Month')
```

```python
plt.figure(figsize=(12, 8))
plt.bar(df['Month'], df['Profit'], color='tab:blue')
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

It seem that July brings the most Profit.

Clothing gives the highest profit.

### Let's see what the men are buying.

```python
# Filter data by selecting only the rows where customer gender is male
male_df = df[df['Customer_Gender'] == 'M']

# Group data by product category and sum the order quantity for each category
male_sales_by_category = male_df.groupby('Product_Category')['Order_Quantity'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(12,8))
sb.barplot(x='Product_Category', y='Order_Quantity', data=male_sales_by_category, color='blue')
plt.title('Total Quantity Sold to Male Customers by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
male_df = df[df["Customer_Gender"] == "M"]
male_top_products = male_df.groupby(["Customer_Gender", "Product_Category"])["Order_Quantity"].sum().reset_index().sort_values("Order_Quantity", ascending=False).head(10)
print(male_top_products)
```

Looks like the men are buying accessories the most here.

### Let's find out how much each country contribute to the company's profit.

```python
profit_country = df.groupby('Country')['Profit'].sum().reset_index()
plt.figure(figsize=(12, 8))
ax = sb.barplot(x="Country", y="Profit", data=profit_country, palette=sb.color_palette("husl", n_colors=len(profit_country)))
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.title('Total Profit by Country')
plt.show()
```

```python
# Group the data by Product_Category and Country and get the sum of Order_Quantity
grouped_data = df.groupby(['Product_Category', 'Country'])['Order_Quantity'].sum().reset_index()

# Filter the data for United States and get the row with the highest Order_Quantity
us_data = grouped_data[grouped_data['Country'] == 'United States']
max_quantity = us_data['Order_Quantity'].max()
max_category = us_data[us_data['Order_Quantity'] == max_quantity]['Product_Category'].iloc[0]

# Plot a bar chart of the data
plt.figure(figsize=(12, 8))
ax = sb.barplot(x="Product_Category", y="Order_Quantity", data=us_data, palette=sb.color_palette("husl", n_colors=len(us_data)))
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.title('Highest Quantity Product Categories in United States')
plt.xlabel('Product Category')
plt.ylabel('Order Quantity')
plt.show()
```

```python
# Group the data by Product_Category and Country and get the sum of Order_Quantity
grouped_data = df.groupby(['Product_Category', 'Country'])['Profit'].sum().reset_index()

# Filter the data for United States and get the row with the highest Order_Quantity
us_data = grouped_data[grouped_data['Country'] == 'United States']
max_quantity = us_data['Profit'].max()
max_category = us_data[us_data['Profit'] == max_quantity]['Product_Category'].iloc[0]

# Plot a bar chart of the data
plt.figure(figsize=(12, 8))
ax = sb.barplot(x="Product_Category", y="Profit", data=us_data, palette=sb.color_palette("husl", n_colors=len(us_data)))
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.title('Highest Profit by Product Categories in United States')
plt.xlabel('Product Category')
plt.ylabel('Profit')
plt.show()
```

Find the distribution of Profit contributed by each product category.

```python
df['Product_Category'].unique()
```

```python
categories = ['Accessories', 'Bikes', 'Clothing']
df_filtered = df[df['Product_Category'].isin(categories)]
# Calculate the total profit for each product category
profit_by_category = df_filtered.groupby('Product_Category')['Profit'].sum().reset_index()

# Set the color palette for the pie chart
colors = sb.color_palette('pastel')[0:len(profit_by_category)]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(profit_by_category['Profit'], labels=profit_by_category['Product_Category'], colors=colors, autopct='%1.1f%%')
plt.title('Total Profit by Product Category')
plt.show()
```
